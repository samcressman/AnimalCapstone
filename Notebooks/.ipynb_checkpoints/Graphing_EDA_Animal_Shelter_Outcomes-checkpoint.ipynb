***Graphing/EDA Notebook***

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
animals_graphing = pd.read_csv("../CSVs/cleaned_animals_pre_numeric_pre_dummies.csv")
```

```python
# Quick check

animals_graphing.head();
```

```python
# Changing back to DateTimes

animals_graphing["Intake Time"] = pd.to_datetime(animals_graphing["Intake Time"]).dt.date
animals_graphing["Outcome Time"] = pd.to_datetime(animals_graphing["Outcome Time"]).dt.date
animals_graphing["Date of Birth"] = pd.to_datetime(animals_graphing["Date of Birth"]).dt.date
```

***Intake Animal Types***

```python
animal_type_count = animals_graphing["Animal Type"].value_counts().reset_index()

plt.figure(figsize=(8, 8))

plt.pie(animal_type_count["Animal Type"].values, labels = animal_type_count["index"].values, autopct = "%0.1f%%")

plt.axis("equal")

plt.title("Intakes: mainly dogs and cats", loc = "right")

plt.show()
```

```python
animal_type_count = animals_graphing["Animal Type"].value_counts(normalize = True).reset_index()

animal_type_count.plot(kind = "barh", x = "index", y = "Animal Type", legend = False, color = "blue")

order = animals_graphing["Outcome Subtype"].value_counts().index

plt.show()
```

***Outcome Type***

```python
# Overall outcome types

outcome_counts = animals_graphing.groupby("Outcome Type")["Animal Type"].count().reset_index()

plt.figure(figsize=(8, 8))

plt.pie(outcome_counts["Animal Type"].values, labels = outcome_counts["Outcome Type"].values, autopct = "%0.1f%%")

plt.title("Overall Austin shelter animals fare well!", loc = "left")

plt.axis("equal")

plt.show()
```

```python
# Overall outcome types

plt.figure(figsize=(10, 10))

plt.title("Overall Austin shelter animals fare well!")

sns.countplot(x = "Animal Type", data = animals_graphing, hue = "Outcome Type")

plt.legend(loc = 7)

plt.show()
```

```python
# Dog outcomes

mask = animals_graphing[animals_graphing["Animal Type"] == "Dog"]

mask = mask.groupby("Outcome Type")["Animal Type"].count().reset_index()

plt.figure(figsize=(6, 6))

plt.pie(mask["Animal Type"].values, labels = mask["Outcome Type"].values, autopct = "%0.1f%%")

plt.title("Dogs fare well!", loc = "left")

plt.axis("equal")

plt.show()
```

```python
# Cat outcomes

mask = animals_graphing[animals_graphing["Animal Type"] == "Cat"]

mask = mask.groupby("Outcome Type")["Animal Type"].count().reset_index()

plt.figure(figsize=(6, 6))

plt.pie(mask["Animal Type"].values, labels = mask["Outcome Type"].values, autopct = "%0.1f%%")

plt.title("Cats fare well!", loc = "left")

plt.axis("equal")

plt.show()
```

```python
# Other

mask = animals_graphing[animals_graphing["Animal Type"] == "Other"]

mask = mask.groupby("Outcome Type")["Animal Type"].count().reset_index()

plt.figure(figsize=(6, 6))

plt.pie(mask["Animal Type"].values, labels = mask["Outcome Type"].values, autopct = "%0.1f%%")

plt.title('"Other" animals (ex. bats, raccoons) are not as lucky')

plt.axis("equal")

plt.show()
```

***Outcome subtype***

```python
# Half null

animals_graphing["Outcome Subtype"].isnull().sum();
```

```python
animals_graphing["Outcome Subtype"].value_counts();
```

```python
# Examing Outcome subtype: majority of animals are given to a partner or find a foster home

# SCRP: Austins Stray Cat Return Program

plt.figure(figsize=(25, 10))
plt.title("Austin appears to work very well with breed-specific partners and foster homes")
sns.countplot(x = "Outcome Subtype", data = animals_graphing, color = "c", 
             order = animals_graphing["Outcome Subtype"].value_counts().index)

plt.show()
```

***Intake Type vs. Outcome Type***

```python
plt.figure(figsize=(10, 10))

plt.title("Most intakes are strays, and most intakes find a successful outcome")

sns.countplot(x = "Intake Type", data = animals_graphing, hue = "Outcome Type")

plt.legend(loc = 7)

plt.show()
```

***Name***

```python
# Most animals do not enter the shelter system with a name, and many are not given a name even upon outcome

plt.figure(figsize=(8, 8))

plt.title("Most animals leave the shelter system with a name")

sns.countplot(x = "has_name", data = animals_graphing, color = "c", 
             order = animals_graphing["has_name"].value_counts().index)

plt.show()
```

```python
# Give every animal a name!

mask = animals_graphing[(animals_graphing["has_name"] == 1)]

plt.figure(figsize=(12, 12))

plt.title("Animals with names tend to have more successful outcomes")

sns.countplot(x = "Outcome Type", data = mask, color = "c", 
             order = mask["Outcome Type"].value_counts().index)

plt.show()
```

```python
# Give every animal a name!

mask = animals_graphing[(animals_graphing["has_name"] == 0)]

plt.figure(figsize=(12, 12))

plt.title("Animals without names have a much higher chance of euthanasia")

sns.countplot(x = "Outcome Type", data = mask, color = "c", 
             order = mask["Outcome Type"].value_counts().index)
```

***Sex upon Outcome***

```python
# Plot outcome based on sex

plt.figure(figsize=(10, 10))

plt.title("Fixed animals are much more likely to be adopted")

sns.countplot(x = "Outcome Type", data = animals_graphing, hue = "Sex upon Outcome")

plt.show()
```

***Intake/Outcome Time, Adoption vs. Euthanasia***

```python
# Intake time

plt.figure(figsize=(20, 10))

plt.title("Seasonality: Austin shelters take in more animals during the middle of the year")

animals_graphing["Intake Condition"].groupby(animals_graphing["Intake Time"]).count().plot(kind="line")

plt.show()
```

```python
# Outcome Type: Adoption

mask = animals_graphing[animals_graphing["Outcome Type"] == "Adoption"]

plt.figure(figsize=(20, 10))

plt.title("Shelter outcomes: annual Clear the Shelters Adoption Event in August")

mask["Outcome Type"].groupby(mask["Outcome Time"]).count().plot(kind="line")

plt.show()
```

```python
# Outcome Type: Euthanasia

mask = animals_graphing[animals_graphing["Outcome Type"] == "Euthanasia"]

animals_graphing["Outcome Time"] = animals_graphing["Outcome Time"]

plt.figure(figsize=(20, 10))

plt.title("Euthanasia seems to spike near the beginning of each year. March 2016: bats!")

mask["Outcome Type"].groupby(mask["Outcome Time"]).count().plot(kind="line")

plt.show()
```

```python
# Trying to explain spike in early 2016: much higher intake of wildlife, mainly bats

for i in range(1, 5):

    mask = animals_graphing[(animals_graphing["Outcome Type"] == "Euthanasia") &
                            (animals_graphing["Outcome Year"] == 2016) & 
                            (animals_graphing["Outcome Month"] == i)]
    
    print("2016 Month ", i)
    print(mask["Intake Type"].value_counts())
    print(" ")
```

***Euthanasia***

```python
# Other animals make up 6% of all intakes but account for 54% of euthanizations

animals_graphing["Animal Type"].value_counts(normalize = True);

# Dog      0.539148
# Cat      0.394134
# Other    0.066718
```

```python
# 8% of all animals are euthanized

mask = animals_graphing[animals_graphing["Outcome Type"] == "Euthanasia"]

mask.shape; # (6334, 128)

mask.shape[0] / animals_graphing.shape[0] # Percent of animals euthanized: 0.08391404573275747

mask["Animal Type"].value_counts(normalize = True);

# Other    0.538207
# Cat      0.240448
# Dog      0.221345
```

```python
# 6157 / 6334 euthanized animals contain an Outcome Subtype

sum(mask["Outcome Subtype"].value_counts()) # 6157

plt.figure(figsize=(11, 11))

sns.countplot(x = "Outcome Subtype", data = mask, color = "c", 
             order = mask["Outcome Subtype"].value_counts().index)

plt.title("Rabies risk and suffering account for the vast majority of euthanasias")

plt.show()
```

```python
# Examing euthanized breeds: bats, cats, raccoons, pit bulls, opossums...

mask["Breed"].value_counts(normalize = True);
```

```python
mask_breed_count = mask["Breed"].value_counts(normalize = True).reset_index().head(15)
mask_breed_count;
```

```python
# Plotting top features

breed_features_top = mask_breed_count.sort_values("Breed", ascending = False).head(15)

breed_features_top = mask_breed_count.sort_values("Breed", ascending = True)

breed_features_top.plot(kind = "barh", x = "index", y = "Breed", legend = False, color = "blue")

plt.xlabel("Percentage")
plt.ylabel("Breed")
plt.title("Bats and shorthair cats represent nearly half of euthanasias")

plt.show()
```
